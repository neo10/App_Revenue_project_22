 ## Missing values
 This function i wrote, checks if the length if an sublist is not equal the length of the header sublist.
 It returns a list, with all index'es where the length is not how it should be. 
 This implies, that there is data missing, or too much datapoints.
 To check that, you'll have to look at that index manually.
 I see that in the AppleStore Dataset is something missing at Index 10473.
 I delete the whole row. 

```python
def list_length_check(dataset):
    header_length=len(dataset[0])
    missing_values_index=[]
    for counter, value in enumerate(dataset):
        if len(value)!=header_length:
            missing_values_index.append(counter)
        
    if missing_values_index==[]:
        print("Check! All good!")
    else: print("Length not correct at index: "+str(missing_values_index))
                                
#---------------------------------------------------------------------

print("Google:") #Fehler bei Index 10473
list_length_check(apps_data_google)
print("\n")
print("Apple:") #Alles gut
list_length_check(apps_data_apple)

del apps_data_google[10473] # Löschen der Fehlerhaften Zeile !ONLY RUN THIS ONCE! 
del apps_data_google[10200]        
```

# Remove all non-roman letters, the way Dataquest teaches us
Since my take above worked, but filtered out too much, so my function wasnt really useful, i looked at the Dataquest
solution and rebuild it. My problem above was, that my if clause didnt count how many characters at any title were out of the ascii range. So even if only 1 "®" was there, the whole datapoint was marked as faulty. 
Now above i will iterate over every character of the title string. When there are more than 3 characters out of the interval 0 - 127 in ASCII code, the title will be deleted. Key function is the ord() function, which returns the ASCII number of a character. **google_clean -> clean_roman | apple_clean -> clean_roman**

```python
def roman_letters_only(dataset,title_column):
    titles=[value[title_column] for value in dataset] # extrahiere nur die "App" Column um nicht das ganze Dataset zu "for loopen index bleibt gleich"
    global clean_roman
    clean_roman=[]
    for i, string in enumerate(titles):        
        x=0
        for char in string:                
            if ord(char)<127 and ord(char)>0:
                x+=1
            if x>len(string)-4:
                clean_roman.append(dataset[i])
                break

roman_letters_only(google_unique,0)
google_unique_roman=clean_roman
roman_letters_only(apple_unique,1)
apple_unique_roman=clean_roman

print("Length GOOGLE: ", len(google_unique_roman))
print("Length APPLE: ", len(apple_unique_roman))

print(apple_unique_roman[:3])
```

# Install Value Dictionary
To eazily apply thresholds on my Datasets, i wrote this code, to create a dictionary, which for example returns float 1 000 000 for the key "1M"

```python
liste=[]
installs_dict={}
for data in google_unique_roman_free[1:]:
    if data[5] not in liste:
        liste.append(data[5])

liste.sort(reverse=True)
for value in liste:
    value=value.replace("+","")
    value=value.replace(",","")
    if "000000000" in value:
        value2=value.replace("000000000","")
        
        installs_dict[value2+"B"]=float(value)
    elif "000000" in value:
        value2=value.replace("000000","")
        
        installs_dict[value2+"M"]=float(value)
        
    elif "000" in value:
        value2=value.replace("000","")
        
        installs_dict[value2+"k"]=float(value)
    else:
        
        installs_dict[value]=float(value)

print(installs_dict)

print(installs_dict["1M"])
```

# Visualize our data
Now we've cleaned our data. Now i create a frequency tables of the "Genres" / "prime_genre" of each dataset.  
In easy words: **Which app Genre has the most apps with high download values.** 
Then we'll see which genre is most succesfull in each App-Store. 

For our next steps, we will know, what genre our new app should be in, to have a higher success probability. 
If our new app is profitable in the google Playstore after six months, we'll code the same app for the AppStore. 
Google "Genres" Column Index: 9
Apple "prime_genre" Column Index: 11

```python
##my own function##
def frequency_table(dataset,index):
    freq_dict={}
    l=len(dataset)-1 # Header -1
    for data in dataset[1:]:
        genre=data[index]
        if genre in freq_dict:
            freq_dict[genre]+=1
        else:
            freq_dict[genre]=1
    for key, value in freq_dict.items():
        freq_dict[key]=round((value/l)*100,2) # dict values in percent
        
    return freq_dict






##this function is given from Dataquest##
def display_table(dataset, index):
    table = frequency_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])

print("Google Categorys :")
display_table(google_unique_roman_free,1)
print("\n Google Genres :")
display_table(google_unique_roman_free,9)
print("\n Apple Prime_Genres :")
display_table(apple_unique_roman_free,11)
```
